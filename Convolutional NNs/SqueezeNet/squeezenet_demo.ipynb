```
import numpy as np
import os, sys
from keras.applications.imagenet_utils import preprocess_input, decode_predictions
from keras.preprocessing import image
```

```
import keras.backend as K
import matplotlib.pyplot as plt
%matplotlib inline
```

```
cwd='/content/gdrive/My Drive/Colab Notebooks/Squeezenet/images/cat.jpeg'
img = image.load_img(cwd, target_size=(227, 227))
```

```
plt.imshow(img)
```

```
sys.path.append('/content/gdrive/My Drive/Colab Notebooks/Squeezenet/keras_squeezenet')
```

```
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```
from squeezenet import SqueezeNet
```

```
model = SqueezeNet(weights=None)
```

```
W = K.eval(model.weights[0])
print(W.shape)
plt.figure(figsize=(18,12))
for idx in range(W.shape[-1]):
    plt.subplot(8,8,idx+1)
    plt.imshow(W[:,:,:,idx]); plt.colorbar()
```

```
preds = model.predict(x)
print('Predicted:', decode_predictions(preds))
# These predictions should look random (since the network is not trained)! 
```

```
model = SqueezeNet(weights='imagenet')
```

```
preds = model.predict(x)
print('Predicted:', decode_predictions(preds))
#Tabby is a cat
```
